#### logistic regression

```r
# Weighted
start_time <- Sys.time()
#
logit <- lapply(phecodes, function(phecode) {
  formula_str <- paste("HS ~ `", phecode, "`+ Asian + Black.or.African.American + More.than.one.population + 
No.answer.race + Other.race + White + Female + Male + Other.sex + Hispanic.or.Latino + No.answer.ethinicity +
Not.Hispanic.or.Latino + Other.ethinicity + age_normalized", sep="")
  model <- glm(formula=as.formula(formula_str), data = matrix, family = binomial, 
               weights = matrix$weight) #
  summary(model)
})
#
end_time <- Sys.time()

# Calculate the time difference
time_taken <- end_time - start_time
print(time_taken)
```

```r
# Estimate
beta_vals <- unlist(lapply(logit, function(res) res$coefficients[2, 1]))
std_vals <- unlist(lapply(logit, function(res) res$coefficients[2, 2]))
z_vals <- unlist(lapply(logit, function(res) res$coefficients[2, 3]))                       
# p-value
p_values_intercept <- unlist(lapply(logit, function(res) res$coefficients[2, 4]))
adjusted_pvalues_bh <- p.adjust(p_values_intercept, method = "BH")  # "BH" is the alias for Benjamini-Hochberg
```

```r
#
plot_data <- data.frame(
  Phecode = phecodes, 
  Beta = beta_vals,
  Std = std_vals,
  Z_score = z_vals,
  Pvalue = p_values_intercept,
  Adjust_PV = adjusted_pvalues_bh
)
plot_data$NegativeLogP = -log10(plot_data$Adjust_PV)
plot_data$BetaDirection <- ifelse(plot_data$Beta > 0, "Positive", "Negative")
```

```r
Phecode_category <- read.csv("icd2phe.csv",colClasses = "character")
# Merge Category
mapping_df <- Phecode_category[, c("Phecode","Phenotype" ,"category")]
mapping_df$Phecode <- paste("X", mapping_df$Phecode,sep="")
mapping_df <- mapping_df[!duplicated(mapping_df$Phecode), ]
```

```r
plot_data_merge <- merge(plot_data, mapping_df, by = "Phecode", how = 'left')
```

```r
plot_data_merge
```

```r
write.csv(plot_data, "plot_data_df.csv", row.names = FALSE)
```
